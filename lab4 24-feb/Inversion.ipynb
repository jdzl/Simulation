```python
from randomG import Cm
import numpy as np

def GenVAD(N,x,px):
    
    X = []
    u =  Cm(N)    
    Px = np.cumsum(px)    
    print u
    print "------"
    for index in range(N):
        for i in range (1,len(Px)):
           
            if(u[index]<Px[0] ):
                X.append(x[0])
                break
                
            if(u[index]<Px[i] and u[index]>=Px[i-1]  ):                
                X.append(x[i-1])
                break
            
                
    return X


x = np.arange(1,7)
px = np.ones(6)*(float(1)/int(6))    
N=1000
print GenVAD(N,x,px)


    
    
    
```
